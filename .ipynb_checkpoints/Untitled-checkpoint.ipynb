```python
#FUNCTIONS

#______________________________________________________________#

def getList(foldername):
    import glob
    dataset = [f for f in glob.glob(foldername+"/*.txt")]
    return (dataset)


#______________________________________________________________#

#with name of file
def inverted_index(corpus):
    import nltk
    from nltk.tokenize import sent_tokenize , word_tokenize
    nltk.download('punkt')
    from collections import defaultdict    
    from nltk.stem import PorterStemmer
    from nltk.corpus import stopwords
    import re

    inverted_index = defaultdict(set)
    stwords = set(stopwords.words('english'))
    ps = PorterStemmer()

    s=getList("testdata")

    for file in s:
        print(file)
        fname = file
        file = open(file , "r")
        text = file.read()
        #id=num
        #num+=1
        text=remove_special_characters(text)
        for sent in sent_tokenize(text):
            for word in word_tokenize(sent):
                word_lower = word.lower()
                if word_lower not in stwords:
                    word_stem = ps.stem(word_lower)
                    # We add the document to the set againt the word in our
                    # index
                    n=len(word_stem)
                    word_stem=word_stem+'$'

                    while(n>=0):
                        word_stem=rotate(word_stem,1)
                        inverted_index[word_stem].add(fname)
                        n-=1


    return inverted_index



#______________________________________________________________#

def getClosest(word):
    dictionary = {}
    for w in inverted_index.keys():
        dist=editDist(word,w)
        dictionary[w]=dist
        if dist == 1:
            return w
    v=(sorted(dictionary.items(), key=lambda item: item[1]))
    return v[0][0]
#______________________________________________________________#

#______________________________________________________________#

def editDist(str1, str2):
    m=len(str1)
    n=len(str2)
    # Create a table to store results of subproblems
    dp = [[0 for x in range(n + 1)] for x in range(m + 1)]
 
    # Fill d[][] in bottom up manner
    for i in range(m + 1):
        for j in range(n + 1):
 
            # If first string is empty, only option is to
            # insert all characters of second string
            if i == 0:
                dp[i][j] = j    # Min. operations = j
 
            # If second string is empty, only option is to
            # remove all characters of second string
            elif j == 0:
                dp[i][j] = i    # Min. operations = i
 
            # If last characters are same, ignore last char
            # and recur for remaining string
            elif str1[i-1] == str2[j-1]:
                dp[i][j] = dp[i-1][j-1]
 
            # If last character are different, consider all
            # possibilities and find minimum
            else:
                dp[i][j] = 1 + min(dp[i][j-1],        # Insert
                                   dp[i-1][j],        # Remove
                                   dp[i-1][j-1])    # Replace
 
    return dp[m][n]
#______________________________________________________________#



def remove_special_characters(text):
    import re
    regex = re.compile('[^a-zA-Z0-9\s]')
    text_returned = re.sub(regex,'',text)
    return text_returned

# def allpermuation(inverted_index):
#     allperm_index=defaultdict(set)
#     for key in inverted_index.keys():
#         n=len(key)
#         perm=key+'$'
#         allperm[perm]=inverted_index[key].values()
#         while(n--):
#             perm=rotate(perm,n)
#             allperm[perm]=inverted_index[key].values()
        

#______________________________________________________________#
        
def rotate(strg, n):
    return strg[n:] + strg[:n]

#______________________________________________________________#
```

```python
query = input('Enter your query:')
query = query.lower()
query_tokens=query.split()

connecting_words = []
# cnt = 1
different_words = []


from nltk.stem import PorterStemmer
ps=PorterStemmer()

from collections import deque
from functions import *
import re

def getClosest(word):
    '''
    function to find closest word in vocabulary
    '''
    dictionary = {}
    for w in ii.keys():
        dist=editDist(word,w)
        dictionary[w]=dist
        if dist == 1:
            return w
    v=(sorted(dictionary.items(), key=lambda item: item[1]))
    return v[0][0]

# Initializing a queue
q = deque()

for word in query_tokens:
    word=ps.stem(word)
    
    if word.lower() != "and" and word.lower() != "or" and word.lower() != "not":
        word=word+'$'
        if '*' in word:
            while(word[-1]!='*'):
                word=rotate(word,1)
        different_words.append(word.lower())
        if len(connecting_words)>0 and connecting_words[-1]=="not":
            if word in ii.keys():
                set_a=ii[word]
                negated_set=(set(dataset).difference(set(set_a)))
                q.append(negated_set)
            else:
                cword=getClosest(word)
                set_a=ii[cword]
                negated_set=(set(dataset).difference(set(set_a)))
                q.append(negated_set)
            continue
        if word in ii.keys():
            q.append(ii[word])
        else:
            cword=getClosest(word)
            q.append(ii[cword])
    else:
        connecting_words.append(word.lower())

print(connecting_words)
print(different_words)


allfiles=getList("testcase")

for connector in connecting_words:
    if connector == 'not':
        continue
#         set_a= q.pop()
#         q.appendleft(set(dataset).difference(set(set_a)))
    elif connector == 'and':
        set_a = q.pop()
        set_b = q.pop()
        intersected=set(set_a).intersection(set(set_b))
        q.appendleft(intersected)
    elif connector == 'or':
        set_a = q.pop()
        set_b = q.pop()
        unioned=set(set_a).union(set(set_b))
        q.appendleft(unioned)

outputset=q.pop()
print(outputset)
print(len(outputset))
```
